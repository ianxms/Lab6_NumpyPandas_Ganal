```python
#Try it 10
df_kaggle = pd.read_csv('survey_results_public.csv')
print("Loaded Dataset:\n", df_kaggle.head())

df_subset = df_kaggle[['Country', 'EdLevel', 'YearsCodePro', 'ConvertedCompYearly']]
print("Subset of Data:\n", df_subset.head())

# 1. Replace the text strings with numbers
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].replace({'Less than 1 year': 0, 'More than 50 years': 51})

# 2. Convert the whole column to numbers (integers)
df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'])

# 3. NOW run your logic
df_clean['ExperienceLevel'] = np.where(df_clean['YearsCodePro'] >= 10, 'Senior', 'Junior')

# Group by the new experience level and get the average salary
grouped_data = df_clean.groupby('ExperienceLevel')['ConvertedCompYearly'].mean().reset_index()
print("Formatted Grouped Data:\n", grouped_data)

# Grouping by Education Level and computing the median salary
grouped_edu = df_clean.groupby('EdLevel')['ConvertedCompYearly'].median().reset_index()

# Renaming the column for clarity
grouped_edu.columns = ['Education Level', 'Median Salary']
print("Median Salary by Education Level:\n", grouped_edu)

# Example for Standard Deviation
std_edu = df_clean.groupby('EdLevel')['ConvertedCompYearly'].std().reset_index()
print("\nSalary Standard Deviation by Education Level:\n", std_edu)

# Group by Country and calculate average salary
country_stats = df_clean.groupby('Country')['ConvertedCompYearly'].mean().reset_index()

# Sort by salary in descending order and take the top 10
top_10_countries = country_stats.sort_values(by='ConvertedCompYearly', ascending=False).head(10)

print("Top 10 Countries with Highest Average Compensation:\n", top_10_countries)
```
